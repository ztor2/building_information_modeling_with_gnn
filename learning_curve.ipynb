```python
import numpy as np
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
elements = ['train_accs', 'train_losses', 'val_accs', 'val_losses']
objects = []
for i in elements:
    with open('{}.pkl'.format(i), 'rb') as f:
        objects.append(pkl.load(f))
```

```python
epochs= range(200) # num_epochs from train process
```

```python
plt.subplots(1, figsize=(7,5), dpi=100)
plt.plot(epochs, objects[0], color="b", label="train_acc")
plt.plot(epochs, objects[1], color='r',  label="train_loss")
# plt.plot(epochs, objects[2], color="b", label="val_acc")
# plt.plot(epochs, objects[3], color="r", label="val_loss")
plt.title("GCN Learning Curve(Cora)", pad=20, size=12)
plt.xlabel("Epoch"), plt.ylabel("Score"), plt.legend(loc="best")
plt.tight_layout()
plt.show()
```
